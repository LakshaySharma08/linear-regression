```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X - 5) ** 2
print(X, Y)
```

 Given a Function f(x), we want to find the value of x that minimises f

```python
plt.style.use('seaborn')
plt.plot(X,Y)
plt.ylabel("F(x)")
plt.xlabel("X")
plt.show()
```

```python
x = 0
lr = 0.1
error = []
#Take 50 steps in downward direction
for i in range(50):
    grad = 2 * (x - 5)
    x = x - lr*grad
    y = (x - 5) ** 2
    error.append(y)
    plt.scatter(x, y)
    print(x)
```

```python
#plot the values of error
plt.plot(error)
```
